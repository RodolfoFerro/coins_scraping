# Using __Beautifoul soup__ to extract web data

## We import our libs:

```python
# Importing our libs:
from IPython.display import display
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

## We set the url:

```python
# Setting path:
#url = "http://numismatics.org/ocre/results?q=imagesavailable%Atrue"
url = "http://numismatics.org/ocre/results?q=&start=0"

r    = requests.get(url)
data = r.text
soup = BeautifulSoup(data)

# Print a 'beautiful' verion of file:
print(soup.prettify)
```

## We look for our data of interest:

```python
results = soup.find_all("div", class_="result-doc")
display(results[0])
```

```python
tags = results[0].find_all("dt")
display(tags)
tags = [tags[i].text for i in range(len(tags))]
display(tags)
```

```python
df = pd.DataFrame(columns=None)

for result in results:
    data = result.find_all("dd")
    df.append(data, ignore_index = True)

display(df.head())
```
