# Using __OpenCV__ to detect image features

## Importing our libs

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Loading images

```python
heads  = cv2.imread('images/heads01111.jpg', 0)
tails  = cv2.imread('images/tails01111.jpg', 0)
```

## Applying the Canny filter

Canny filter is a gradient filter that allows us to detect the edges of an object contained in an image.

Check this links for more info:
* [Canny edge detector (from Wikipedia)](https://en.wikipedia.org/wiki/Canny_edge_detector)
* [Canny Edge Detection (an OpenCV example)](http://docs.opencv.org/trunk/da/d22/tutorial_py_canny.html)

```python
# As simple as this:
hedges = cv2.Canny(heads, 100, 200)
tedges = cv2.Canny(tails, 100, 200)
```

## Plotting the results

```python
# Heads and edges:
plt.subplot(221), plt.imshow(heads, cmap = 'gray')
plt.title('Original Heads'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(hedges, cmap = 'gray')
plt.title('Edge Heads'), plt.xticks([]), plt.yticks([])

# Tails and edges
plt.subplot(223), plt.imshow(tails, cmap = 'gray')
plt.title('Original Tails'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(tedges, cmap = 'gray')
plt.title('Edge Tails'), plt.xticks([]), plt.yticks([]);
```

## What's next?

We created a utility function to automatically detect the edges of all images, this can be found es `EdgeGradient` in the `imgprocessing.py` script.

### Checkout and compare the following example of using some preprocessing functions to obtain data from coins.

```python
from imgprocessing import *

heads   = cv2.imread('images/heads01101.jpg',0)
no_proc = edgeGradient(heads)            # No preprocessing   
gammaad = adjustGamma(heads, gamma=1.8)  # Process contrast
preproc = blurImage(gammaad, 2)          # We can play with blur kernels 
preproc = edgeGradient(preproc) 

plt.figure(figsize=(8,8))
plt.subplot(221), plt.imshow(heads, cmap = 'gray')
plt.title('Original Heads'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(gammaad, cmap='gray')
plt.title('Gamma adjust'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(no_proc, cmap = 'gray')
plt.title('Edge Heads Not Pre-Processed'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(preproc, cmap = 'gray')
plt.title('Edge Heads Preprocessed'), plt.xticks([]), plt.yticks([]);
```

According to this example, a pre-rpecessing before an edge extraction provides better results.
